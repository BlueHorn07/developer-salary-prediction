## LanguageWorkedWith

```python
total_languages = []
```

```python
language2018 = data2018['LanguageWorkedWith']

languages2018 = []

for row in language2018:
  for lang in str(row).split(';'):
    if lang not in languages2018:
      languages2018.append(lang)
    if lang not in total_languages:
      total_languages.append(lang)

print(languages2018)
print(len(languages2018))
```

```python
language2019 = data2019['LanguageWorkedWith']

languages2019 = []

for row in language2019:
  for lang in str(row).split(';'):
    if lang not in languages2019:
      languages2019.append(lang)
    if lang not in total_languages:
      total_languages.append(lang)
      
print(languages2019)
print(len(languages2019))
```

```python
language2020 = data2020['LanguageWorkedWith']

languages2020 = []

for row in language2020:
  for lang in str(row).split(';'):
    if lang not in languages2020:
      languages2020.append(lang)
    if lang not in total_languages:
      total_languages.append(lang)
      
print(languages2020)
print(len(languages2020))
```

```python
lang_overview = {}

print(len(total_languages))

common_languages = []

for lang in total_languages:
  isExist = [0, 0, 0]
  if lang in languages2018:
    isExist[0] = 1
  if lang in languages2019:
    isExist[1] = 1
  if lang in languages2020:
    isExist[2] = 1
  
  if isExist[0] and isExist[1] and isExist[2]:
    common_languages.append(lang)

  lang_overview[lang] = isExist

common_languages.append('HTML')
common_languages.append('CSS')
print(len(common_languages), common_languages)

data = pd.DataFrame(lang_overview).drop('nan', axis=1)
data
```

```python
# process 2018

post_lang2018 = []

for data in language2018:
  row = [0 for _ in common_languages]
  for lang in str(data).split(';'):
    if lang == 'HTML/CSS':
      row[common_languages.index('HTML')] = 1
      row[common_languages.index('CSS')] = 1
    if lang in common_languages:
      row[common_languages.index(lang)] = 1
  post_lang2018.append(row)

data = pd.DataFrame(post_lang2018)
data.columns = common_languages
data = data.drop('nan', axis=1)

final_2018 = pd.concat([final_2018, data],axis=1)
final_2018
```

```python
# process 2019

post_lang2019 = []

for data in language2019:
  row = [0 for _ in common_languages]
  for lang in str(data).split(';'):
    if lang == 'HTML/CSS':
      row[common_languages.index('HTML')] = 1
      row[common_languages.index('CSS')] = 1
    if lang in common_languages:
      row[common_languages.index(lang)] = 1
  post_lang2019.append(row)

data = pd.DataFrame(post_lang2019)
data.columns = common_languages
data = data.drop('nan', axis=1)

final_2019 = pd.concat([final_2019, data],axis=1)
final_2019
```

## WebFrameWorkedWith + MiscTechWorkedWith

*2018년도의 "FrameworkWorkedWith"  = 2019 & 2020 년도의 "WebFrameWorkedWith" + "MiscTechWorkedWith"  

```python
total_MiscTech = []

# 2018
miscTech2018 = data2018['FrameworkWorkedWith']

MiscTech2018 = []

for row in miscTech2018:
  for lang in str(row).split(';'):
    if lang not in MiscTech2018:
      MiscTech2018.append(lang)
    if lang not in total_MiscTech:
      total_MiscTech.append(lang)

print(MiscTech2018)
print(len(MiscTech2018))

# 2019
webFrame2019 = data2019['WebFrameWorkedWith']
miscTech2019 = data2019['MiscTechWorkedWith']
merged2019 = []

for i in range(len(webFrame2019)):
  merged2019.append(str(webFrame2019[i]) + ';' + str(miscTech2019[i]))

MiscTech2019 = []

for row in merged2019:
  for lang in str(row).split(';'):
    if lang not in MiscTech2019:
      MiscTech2019.append(lang)
    if lang not in total_MiscTech:
      total_MiscTech.append(lang)

print(MiscTech2019)
print(len(MiscTech2019))

# 2020
webFrame2020 = data2020['WebframeWorkedWith']
miscTech2020 = data2020['MiscTechWorkedWith']
merged2020 = []

for i in range(len(webFrame2020)):
  merged2020.append(str(webFrame2020[i]) + ';' + str(miscTech2020[i]))

MiscTech2020 = []

for row in merged2020:
  for lang in str(row).split(';'):
    if lang not in MiscTech2020:
      MiscTech2020.append(lang)
    if lang not in total_MiscTech:
      total_MiscTech.append(lang)

print(MiscTech2020)
print(len(MiscTech2020))
```

```python
mistTech_overview = {}

print(len(total_MiscTech))

common_mistTechs = []

for mistTech in total_MiscTech:
  isExist = [0, 0, 0]
  if mistTech in MiscTech2018:
    isExist[0] = 1
  if mistTech in MiscTech2019:
    isExist[1] = 1
  if mistTech in MiscTech2020:
    isExist[2] = 1
  
  if isExist[0] and isExist[1] and isExist[2]:
    common_mistTechs.append(mistTech)

  mistTech_overview[mistTech] = isExist

print(len(common_mistTechs), common_mistTechs)

data = pd.DataFrame(mistTech_overview).drop('nan', axis=1)
data
```

```python
# process 2018
post_misTech2018 = []

for data in miscTech2018:
  row = [0 for _ in common_mistTechs]
  for lang in str(data).split(';'):
    if lang in common_mistTechs:
      row[common_mistTechs.index(lang)] = 1
  post_misTech2018.append(row)

data = pd.DataFrame(post_misTech2018)
data.columns = common_mistTechs
data = data.drop('nan', axis=1)
final_2018 = pd.concat([final_2018, data],axis=1)
final_2018

# process 2019
post_misTech2019 = []

for data in merged2019:
  row = [0 for _ in common_mistTechs]
  for lang in str(data).split(';'):
    if lang in common_mistTechs:
      row[common_mistTechs.index(lang)] = 1
  post_misTech2019.append(row)

data = pd.DataFrame(post_misTech2019)
data.columns = common_mistTechs
data = data.drop('nan', axis=1)
final_2019 = pd.concat([final_2019, data],axis=1)
final_2019

# process 2020
post_misTech2020 = []

for data in merged2020:
  row = [0 for _ in common_mistTechs]
  for lang in str(data).split(';'):
    if lang in common_mistTechs:
      row[common_mistTechs.index(lang)] = 1
  post_misTech2020.append(row)

data = pd.DataFrame(post_misTech2020)
data.columns = common_mistTechs
data = data.drop('nan', axis=1)
final_2020 = pd.concat([final_2020, data],axis=1)
final_2020
```
