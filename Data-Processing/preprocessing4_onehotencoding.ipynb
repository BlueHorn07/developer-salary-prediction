```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
personal = pd.read_csv('personal_noNA.csv')
personal.head(10)
```

```python
personal.info()
```

```python
dummies = pd.get_dummies(personal[['Country', 'Ethnicity', 'Gender', 'UndergradMajor']],
                         columns = ['Country', 'Ethnicity', 'Gender', 'UndergradMajor'],
                        prefix = ['Country', 'Ethnicity', 'Gender', 'UndergradMajor'],
                        drop_first = False)
dummies.columns
```

```python
personal_hot = pd.concat([personal, dummies], axis = 1)
personal_hot.head()
```

```python
personal_hot.to_csv("personal_preprocess.csv",index = False)
```
